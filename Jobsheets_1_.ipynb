```python
import math

def cek_bilangan():
    try:
        # Mengambil input dari pengguna
        angka = int(input("Masukkan sebuah bilangan bulat: "))

        # 1. Cek Genap atau Ganjil
        if angka % 2 == 0:
            status_ganjil_genap = "Genap"
        else:
            status_ganjil_genap = "Ganjil"

        # 2. Cek Bilangan Prima
        is_prima = True
        if angka <= 1:
            is_prima = False
        else:
            # Mengecek pembagi dari 2 hingga akar kuadrat dari angka
            for i in range(2, int(math.sqrt(angka)) + 1):
                if angka % i == 0:
                    is_prima = False
                    break

        status_prima = "Prima" if is_prima else "Bukan Prima"

        # Menampilkan Hasil
        print("-" * 30)
        print(f"Angka: {angka}")
        print(f"Kategori: {status_ganjil_genap}")
        print(f"Status: {status_prima}")
        print("-" * 30)

    except ValueError:
        print("Mohon masukkan angka bulat yang valid.")

# Menjalankan fungsi
cek_bilangan()
```

 Variabel dan Operasi dasar


```python
# --- 1. DEKLARASI VARIABEL ---
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print("=== DATA DIRI ===")
print(f"Nama       : {nama}")
print(f"Umur       : {umur} tahun")
print(f"Tinggi     : {tinggi} cm")
print(f"Mahasiswa  : {is_student}")


# --- 2. OPERASI ARITMETIKA ---
a = 10
b = 3

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b
pembagian = a / b
pembagian_bulat = a // b
modulus = a % b
pangkat = a ** b

print("\n=== OPERASI ARITMETIKA ===")
print(f"Angka a = {a}, b = {b}")
print(f"Tambah (a + b)  : {penjumlahan}")
print(f"Kurang (a - b)  : {pengurangan}")
print(f"Kali   (a * b)  : {perkalian}")
print(f"Bagi   (a / b)  : {pembagian:.2f}")
print(f"Bagi Bulat (//) : {pembagian_bulat}")
print(f"Sisa Bagi (%)   : {modulus}")
print(f"Pangkat (**)    : {pangkat}")


# --- 3. OPERASI PERBANDINGAN ---
print("\n=== OPERASI PERBANDINGAN ===")
print(f"a > b  : {a > b}")
print(f"a < b  : {a < b}")
print(f"a == b : {a == b}")
print(f"a != b : {a != b}")
print(f"a >= b : {a >= b}")
print(f"a <= b : {a <= b}")


# --- 4. OPERASI LOGIKA ---
x = True
y = False

print("\n=== OPERASI LOGIKA ===")
print(f"x = {x}, y = {y}")
print(f"x AND y : {x and y}")
print(f"x OR y  : {x or y}")
print(f"NOT x   : {not x}")


# --- 5. PERCABANGAN (IF-ELSE) ---
print("\n=== LOGIKA PERCABANGAN ===")
if a > b and b > 0:
    print("Hasil: Kondisi terpenuhi (a > b DAN b positif).")
else:
    print("Hasil: Kondisi tidak terpenuhi.")
```

Percabangan

```python
# --- 1. IF SEDERHANA ---
nilai = 85
print("=== IF SEDERHANA ===")
if nilai > 80:
    print("Selamat! Anda lulus dengan nilai tinggi.\n")


# --- 2. IF-ELSE ---
umur = 17
print("=== IF-ELSE ===")
if umur >= 18:
    print("Anda sudah cukup umur untuk mendapatkan SIM.\n")
else:
    print("Anda belum cukup umur untuk mendapatkan SIM.\n")


# --- 3. IF-ELIF-ELSE ---
hari = "Rabu"
print("=== IF-ELIF-ELSE ===")
if hari == "Senin":
    print("Hari Senin - Saatnya kembali bekerja!\n")
elif hari == "Selasa":
    print("Hari Selasa - Jadwal rapat mingguan.\n")
elif hari == "Rabu":
    print("Hari Rabu - Ada diskon di beberapa toko.\n")
else:
    print("Hari lainnya - Atur jadwalmu dengan baik.\n")


# --- 4. IF BERSARANG (NESTED IF) ---
suhu = 35
print("=== NESTED IF ===")
if suhu > 30:
    print("Cuaca cukup panas.")
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air.\n")
    else:
        print("Masih relatif normal, tapi tetap jaga kesehatan.\n")
else:
    print("Cuaca sepertinya cukup sejuk.\n")


# --- 5. PERCABANGAN DENGAN OPERASI LOGIKA ---
nilai_teori = 75
nilai_praktik = 80
print("=== IF DENGAN LOGIKA AND/OR ===")
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik memadai.\n")
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Anda perlu meningkatkan nilai teori dan praktik.\n")
elif nilai_teori < 70:
    print("Anda perlu meningkatkan nilai teori.\n")
else:
    print("Anda perlu meningkatkan nilai praktik.\n")


# --- 6. IF TERNARY (CONDITIONAL EXPRESSION) ---
angka = -5
print("=== IF TERNARY ===")
status = "Positif" if angka > 0 else "Negatif atau Nol"
print(f"Angka = {angka} => {status}")
```

Perulangan


```python
# --- 1. FOR LOOP DENGAN range() ---
print("=== FOR LOOP (range) ===")
for i in range(5):
    print(f"Perulangan ke-{i}")


# --- 2. FOR LOOP MENGITERASI LIST ---
print("\n=== FOR LOOP (list) ===")
buah = ["apel", "mangga", "jeruk", "pisang"]
for item in buah:
    print(f"Buah: {item}")


# --- 3. WHILE LOOP ---
print("\n=== WHILE LOOP ===")
count = 0
while count < 5:
    print(f"count = {count}")
    count += 1


# --- 4. BREAK PADA LOOP ---
print("\n=== BREAK ===")
for i in range(10):
    if i == 3:
        print(f"Loop dihentikan paksa pada i = {i}")
        break
    print(f"i = {i}")


# --- 5. CONTINUE PADA LOOP ---
print("\n=== CONTINUE ===")
for i in range(5):
    if i == 2:
        print(f"Lewati i = {i} dengan continue")
        continue
    print(f"i = {i}")


# --- 6. NESTED LOOP (LOOP BERSARANG) ---
print("\n=== NESTED LOOP ===")
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")


# --- 7. ELSE PADA LOOP ---
print("\n=== ELSE PADA LOOP ===")
for x in range(3):
    print(f"x = {x}")
else:
    print("Loop for selesai sepenuhnya.")

y = 0
while y < 3:
    print(f"y = {y}")
    y += 1
else:
    print("Loop while selesai sepenuhnya.")


# --- 8. PASS SEBAGAI PLACEHOLDER ---
print("\n=== PASS (Placeholder) ===")
for i in range(3):
    if i == 1:
        pass
    print(f"i = {i}")
```
